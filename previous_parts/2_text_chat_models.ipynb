### Model Parameters & difference between "text completion" models and "chat" models

##### General boilerplate code

```python
import langchain
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAI
from langchain_openai import ChatOpenAI, OpenAI
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate

load_dotenv()

google_api_key = os.getenv("GOOGLE_API_KEY")
openai_api_key = os.getenv("OPENAI_API_KEY")


openai_llm_chat = ChatOpenAI(
    temperature=0,
    model="gpt-4o", 
    api_key=openai_api_key,
    max_tokens=30,
    stop="India",
    timeout=30 # in seconds
)

openai_llm_text = OpenAI(
    temperature=0, 
    model="gpt-3.5-turbo-instruct",
    # stop="test",
    api_key=openai_api_key
)

google_llm_chat = ChatGoogleGenerativeAI(
    temperature=0, 
    model="gemini-2.0-flash", 
    api_key=google_api_key,
    # max_tokens=100,
    timeout=0.1,
    )

google_llm_text = GoogleGenerativeAI(
    temperature=0, 
    model="gemini-2.0-flash", 
    api_key=google_api_key,
    max_tokens=30,
    timeout=0.1
    )
    
```

```python
res = google_llm_chat.invoke("Who is Dhoni in 2 lines?")

print(res.content, "\n")
```
