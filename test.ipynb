```python
def add_decorator(func):
    def inner(*args, **kwargs):
        first, second, *rest = args
        if first == 10 or second == 10:
            return func(*args, **kwargs)
        else:
            return "Any one integer should be 10. Try again"
    return inner


@add_decorator
def addition(a:int, b:int) -> int:
    return a + b

addition(45,10)
```
